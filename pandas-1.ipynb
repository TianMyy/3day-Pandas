### 缺失值处理

#### 判断缺失值

```python
df[df['名字'].isnull()]
```

#### 填充缺失值

```python
dit={'名字':'复仇者联盟3', '投票人数':123456, '类型':'剧情/科幻',  '产地':'美国', '上映时间':'2018-05-04 00:00:00', '时长':'142',
     '年代':2018, '评分':np.nan, '首映地点':'美国'}
s=pd.Series(dit)

s.name=38738
df=df.append(s)
```

```python
df[df['评分'].isnull()]
```

```python
df['评分'].fillna(np.mean(df['评分']), inplace=True)
df.tail()
```
